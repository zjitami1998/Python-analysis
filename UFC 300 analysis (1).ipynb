```python
df_filtered = df[['Weight_class', 'height_cm', 'weight_in_kg', 'reach_in_cm', 'win_pct', 'loss_pct']].dropna()

avg_stats_by_weight_class = df_filtered.groupby('Weight_class').agg({'win_pct': 'mean', 'loss_pct': 'mean'})

# Iterate through each weight class and determine the best attribute
for weight_class in avg_stats_by_weight_class.index:
    print(f"Weight Class: {weight_class}")
    avg_win_pct = avg_stats_by_weight_class.loc[weight_class, 'win_pct']
    avg_loss_pct = avg_stats_by_weight_class.loc[weight_class, 'loss_pct']
    print(f"Avg Win Percentage: {avg_win_pct:.2f}%")
    print(f"Avg Loss Percentage: {avg_loss_pct:.2f}%")
    for attribute in ['height_cm', 'weight_in_kg', 'reach_in_cm']:
        avg_attribute = df_filtered[df_filtered['Weight_class'] == weight_class][attribute].mean()
        print(f"Average {attribute}: {avg_attribute:.2f}")
    print("\n")
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Assuming df is your DataFrame with the given columns

# Filter the DataFrame to include only the relevant columns and remove NaN values
df_filtered = df[['Weight_class', 'height_cm', 'weight_in_kg', 'reach_in_cm', 'win_pct']].dropna()

# Calculate the correlation matrix for win percentage for each weight class
correlation_matrices_win = {}
for weight_class, group in df_filtered.groupby('Weight_class'):
    correlation_matrices_win[weight_class] = group[['height_cm', 'weight_in_kg', 'reach_in_cm', 'win_pct']].corr()

# Plot the heatmaps
plt.figure(figsize=(15, 8))

for i, (weight_class, correlation_matrix_win) in enumerate(correlation_matrices_win.items()):
    plt.subplot(2, 3, i+1)
    sns.heatmap(correlation_matrix_win, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title(f'Correlation Heatmap: Win Percentage with Attributes\n(Weight Class: {weight_class})')
    plt.xlabel('Attributes')
    plt.ylabel('Win Percentage')

plt.tight_layout()
plt.show()

```

# Significant Strikes analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming df is your DataFrame with MMA fighter data

# Select relevant columns for significant strikes analysis
significant_strikes_columns = [
    'significant_strikes_landed_per_minute',
    'significant_striking_accuracy',
    'significant_strikes_absorbed_per_minute'
]

# Filter out rows with missing data in significant strikes columns
significant_strikes_data = df[significant_strikes_columns].dropna()

# Distribution plot for significant strikes absorbed per minute
plt.figure(figsize=(8, 6))
sns.histplot(significant_strikes_data['significant_strikes_absorbed_per_minute'], kde=True)
plt.title('Distribution of Significant Strikes Absorbed per Minute')
plt.xlabel('Significant Strikes Absorbed per Minute')
plt.show()
```

```python
# Distribution plot for significant striking accuracy
plt.figure(figsize=(8, 6))
sns.histplot(significant_strikes_data['significant_striking_accuracy'], kde=True)
plt.title('Distribution of Significant Striking Accuracy')
plt.xlabel('Significant Striking Accuracy')
plt.show()
```

```python
# Find the fighter with the most significant strikes landed per minute
max_sla_index = df['significant_strikes_landed_per_minute'].idxmax()
fighter_with_max_sla = df.loc[max_sla_index, 'name']

# Find the name of the fighter with the most significant strikes absorbed per minute
max_ssa_index = df['significant_strikes_absorbed_per_minute'].idxmax()
fighter_with_max_ssa = df.loc[max_ssa_index, 'name']

print("Fighter with the most Significant Strikes Landed per Minute:", fighter_with_max_sla)
print("Fighter with the most Significant Strikes Absorbed per Minute:", fighter_with_max_ssa)
```

```python
grouped_by_weight_class = df.groupby('Weight_class')


max_sla_fighters = {}
max_ssa_fighters = {}


for weight_class, group_df in grouped_by_weight_class:
    # Find the index of the fighter with the highest significant strikes landed per minute in the group
    max_sla_index = group_df['significant_strikes_landed_per_minute'].idxmax()
    max_sla_fighter = group_df.loc[max_sla_index, 'name']
    
    max_sla_fighters[weight_class] = max_sla_fighter
    
    
    max_ssa_index = group_df['significant_strikes_absorbed_per_minute'].idxmax()
    max_ssa_fighter = group_df.loc[max_ssa_index, 'name']
    
    max_ssa_fighters[weight_class] = max_ssa_fighter


print("Fighter with the most Significant Strikes Landed per Minute for each weight class:")
for weight_class, fighter in max_sla_fighters.items():
    print(f"{weight_class}: {fighter}")

print("\nFighter with the most Significant Strikes Absorbed per Minute for each weight class:")
for weight_class, fighter in max_ssa_fighters.items():
    print(f"{weight_class}: {fighter}")
```

# Takedown analysis

```python
# Group the DataFrame by the 'Weight_class' column
grouped_by_weight_class = df.groupby('Weight_class')

# Initialize dictionaries to store the fighters with the highest takedowns landed per minute and takedowns defended per minute for each weight class
max_takedowns_landed_fighters = {}
max_takedowns_defended_fighters = {}

# Iterate over each group
for weight_class, group_df in grouped_by_weight_class:
    # Find the index of the fighter with the highest takedowns landed per minute in the group
    max_takedowns_landed_index = group_df['average_takedowns_landed_per_15_minutes'].idxmax()
    max_takedowns_landed_fighter = group_df.loc[max_takedowns_landed_index, 'name']
    # Store the fighter's name in the dictionary with the weight class as the key
    max_takedowns_landed_fighters[weight_class] = max_takedowns_landed_fighter
    
    # Find the index of the fighter with the highest takedowns defended per minute in the group
    max_takedowns_defended_index = group_df['takedown_defense'].idxmax()
    max_takedowns_defended_fighter = group_df.loc[max_takedowns_defended_index, 'name']
    # Store the fighter's name in the dictionary with the weight class as the key
    max_takedowns_defended_fighters[weight_class] = max_takedowns_defended_fighter

# Print the results
print("Fighter with the most Takedowns Landed per Minute for each weight class:")
for weight_class, fighter in max_takedowns_landed_fighters.items():
    print(f"{weight_class}: {fighter}")

print("\nFighter with the most Takedowns Defended per Minute for each weight class:")
for weight_class, fighter in max_takedowns_defended_fighters.items():
    print(f"{weight_class}: {fighter}")
```
